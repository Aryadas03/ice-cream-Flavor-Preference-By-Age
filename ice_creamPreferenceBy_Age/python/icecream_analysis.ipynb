```python
import pandas as pd
```

```python
df=pd.read_excel("ice_cream_age_flavor.xlsx")
```

```python
df.head()
```

```python
def age_group(age):
    if age <= 12:
        return "Kids"
    elif age <= 19:
        return "Teens"
    elif age <= 40:
        return "Adults"
    else:
        return "Seniors"
```

```python
df["Age_Group"] = df["Age"].apply(age_group)
```

```python
df.head()
```

```python
grouped = df.groupby(["Age_Group", "Flavor"]).size().reset_index(name="Count")
grouped
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=grouped, x="Age_Group", y="Count", hue="Flavor")
plt.title("Ice Cream Flavor Preferences by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of People")
plt.show()
```
